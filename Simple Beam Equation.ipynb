```python
from sympy import *
from sympy.abc import *
```

```python
w = Function("w")(x)
sol = E * I * w.diff(x, 4) + q
```

```python
SS = dsolve(sol,
            ics={
                w.subs(x, 0): 0,
                w.subs(x, L): 0,
                w.diff(x,2).subs(x, 0): 0,
                w.diff(x,2).subs(x, L): 0
            })
W = SS.rhs
W
```

```python
graphSS = plot(W.evalf(subs={
    L: 100,
    E: 210000,
    I: 1 / 12,
    q: 0.1
}), (x, 0, 100))
```
